```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
#Snowball stemming all the words
for _ in range(len(data['complaint'])):
    data['complaint'][_]=" ".join([stemmer.stem(__) for __ in data['complaint'][_].split()])
```

```python
#Is the issue related to Journey
#Finding train : "train between", "station between", "divert between", "from ___ to ___ ", "delay", "water"
#ticket issue : "ticket"
def isJourney(s):
    words=["train between", "divert between", "ticket", "delay", "water", "hours", "late", "hrs"] #TODO: from to
    for word in words:
        if word in s.lower():
            return True
    if "from" in s.lower() and "to" in s.lower():
        return True
    if "between" in s.lower() and "station" in s.lower():
        return True
    return False
```

```python
#Is cleanliness
#Toilet : "clean","cleanli","dirt", "dirti","dust","dusti","smell"
def isClean(s):
    words=["clean","cleanli","dirt", "dirti","dust","dusti","smell"]
    for word in words:
        if word in s.lower():
            return True
    return False
```

```python
#Is food
#food : "bad food","food"
def isFood(s):
    words=["food","lunch","tea","dinner","breakfast"] 
    for word in words:
        if word in s.lower():
            return True
    return False
```

```python
#Is crime and security
#crime : "murder", "illegal"
#security: "inspection", "camera", "scanner"
def isSecurity(s):
    words=["illegal","inspect", "camera","scanner", "help", "alcohol"] 
    for word in words:
        if word in s.lower():
            return True
    return False
```

```python
#Is digital issue
#app : "app", "website", "server", "digital", "technology", "wifi", "internet"
def isDigital(s):
    words=["app", "websit", "server", "digit", "technolog", "wifi", "internet"] 
    for word in words:
        if word in s.lower():
            return True
    return False
```

```python
def isHealth(s):
    words=["health", "medicine", "ill", "hurt", "accident"] 
    for word in words:
        if word in s.lower():
            return True
    return False
```

```python
#classifiying manually
category=[]
for _ in range(len(data['complaint'])):
    s=data['complaint'][_]
    if isJourney(s):
        category.append('journey')
    elif isFood(s):
        category.append('food')
    elif isClean(s):
        category.append('clean')
    elif isDigital(s):
        category.append('digital')
    elif isSecurity(s):
        category.append('security')
    elif isHealth(s):
        category.append('health')
    else:
        category.append('others')
```

```python
# category.count('others'),category.count('food'),category.count('digital'),category.count('journey'),category.count('security'),category.count('clean')
categories = {}
categories['others'] = category.count('others')
categories['food'] = category.count('food')
categories['journey'] = category.count('journey')
categories['clean'] = category.count('clean')
categories['security'] = category.count('security')
categories['health'] = category.count('health')
```
