```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
#Snowball stemming all the words
for _ in range(len(data['complaint'])):
    data['complaint'][_]=" ".join([stemmer.stem(__) for __ in data['complaint'][_].split()])
```

```python
data.head()
```

```python
#Is the issue related to Journey
#Finding train : "train between", "station between", "divert between", "from ___ to ___ ", "delay", "water"
#ticket issue : "ticket"
def isJourney(s):
    words=["train between", "station between", "divert between", "ticket", "delay", "water"] #TODO: from to
    for word in words:
        if word in s:
            return True
    if "from" in s and "to" in s:
        return True
    return False
```

```python
#Is cleanliness
#Toilet : "clean","cleanli","dirt", "dirti","dust","dusti","smell"
def isClean(s):
    words=["clean","cleanli","dirt", "dirti","dust","dusti","smell"]
    for word in words:
        if word in s:
            return True
    return False
```

```python
#Is food
#food : "bad food","food"
def isFood(s):
    words=["food","lunch","tea","dinner","breakfast"] 
    for word in words:
        if word in s:
            return True
    return False
```

```python
#Is crime and security
#crime : "murder", "illegal"
#security: "inspection", "camera", "scanner"
def isSecurity(s):
    words=["illegal","inspection", "camera","scanner"] 
    for word in words:
        if word in s:
            return True
    return False
```

```python
#Is digital issue
#app : "app", "website", "server", "digital", "technology", "wifi", "internet"
def isDigital(s):
    words=["app", "website", "server", "digital", "technology", "wifi", "internet"] 
    for word in words:
        if word in s:
            return True
    return False
```

```python
#classifiying manually
category=[]
for _ in range(len(data['complaint'])):
    s=data['complaint'][_]
    if isJourney(s):
        category.append('journey')
    elif isFood(s):
        category.append('food')
    elif isClean(s):
        category.append('clean')
    elif isDigital(s):
        category.append('digital')
    elif isSecurity(s):
        category.append('security')
    else:
        category.append('others')
```

```python
category.count('others'),category.count('food'),category.count('digital'),category.count('journey'),category.count('security'),category.count('clean')
```

```python
for i in range(len(category)):
    if category[i]=='others':
        print(df['complaint'][i])
```
